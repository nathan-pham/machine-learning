# Polynomial Regression

## Import Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Import Dataset

```python
dataset = pd.read_csv("datasets/position_salary.csv")
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values

# print(np.array(X).reshape(-1, 1))
```

## Create Polynomial Regression Model

```python
from sklearn.preprocessing import PolynomialFeatures

# more degrees = more accuracy (polynomials can represent any function)
poly_reg = PolynomialFeatures(degree=5)
X_poly = poly_reg.fit_transform(X)
```

## Create Linear Regression Model

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)
```

## Visualize Model

```python
# plot the training set
plt.scatter(X, y, color="red")
plt.plot(X, lin_reg.predict(X_poly), color="blue")

# add axis labels
plt.title("Position vs Salary")
plt.xlabel("Position")
plt.ylabel("Salary")

# show the plot
plt.show()
```

### Visualize Model with Smoothing

```python
# create grid
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))

# plot the training set
plt.scatter(X, y, color="red")
plt.plot(X_grid, lin_reg.predict(poly_reg.fit_transform(X_grid)), color="blue")

# add axis labels
plt.title("Position vs Salary (Smooth)")
plt.xlabel("Position")
plt.ylabel("Salary")

# show the plot
plt.show()
```
